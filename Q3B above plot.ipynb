```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x =np.arange(0,100)
y =x*2
z =x**2
```

```python
fig = plt.figure()
ax1 =fig.add_axes([0,0,1,1])
ax2 =fig.add_axes([0.2,0.5,0.4,0.4])

ax1.plot(x,z)
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax1.set_title('z versus x')

ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('y versus x')
```
